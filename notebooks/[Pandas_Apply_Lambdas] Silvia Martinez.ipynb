## Challenge 1. Using a single argument

```python
def bins(votes):
    if votes >= 0 and votes <= 999:
        return "cat_1"
    elif votes >= 1000 and votes <= 9999:
        return "cat_2"
    elif votes >= 10000 and votes <= 99999:
        return "cat_3"
    elif votes >= 100000 and votes <= 999999:
        return "cat_4"
    elif votes >= 1000000:
        return "cat_5"
        
bins(1000)
```

```python
df["Bin"] = df.apply(lambda x: bins(x["Votes"]), axis = 1)
```

```python
df
```

## Challenge 2. Using 2 arguments

```python
def revenue_minute(revenue, minute):
    return revenue / minute
```

```python
df["Revenue per Minute"] = df.apply(lambda x: revenue_minute(x["Revenue (Millions)"], x["Runtime (Minutes)"]), axis = 1)
```

## Challege 5. And finally some fantasy...

```python
# I am goin to establish the average rating taking in consideration
#web rating(metascore) and user rating(rating).
```

```python
#I am going to convert the metascore rating into a 1-10 value, to be equal to the other rating.

df["Metascore"] = df["Metascore"]/10
```

```python
df
```

```python
df["Average_Rating"] = (df["Rating"] + df["Metascore"]) / 2
```

```python
df
```

```python
top_avg_rated_movies = df.sort_values(by="Average_Rating", ascending=False)
```

```python
top_avg_rated_movies
```

```python
# I am going to create a new ranking taking in consideration the number of votes. If a film has more than 30000 votes, the ranking will decrease 1 point. It will add 1 point to the ones below that number of votes.

def ranking_votes(votes, rating):
    if votes >= 30000:
        rating = rating - 1
    else:
        rating = rating + 1
    return rating

df["Rating_Votes"] = df.apply(lambda x: ranking_votes(x["Votes"], x["Rating"]), axis = 1)
```

```python
df
```

```python
#Finally, I am going to create a rating based on the the other kind of ratings. 

df["Final_Rating"] = (df["Rating"] + df["Metascore"] + df["Average_Rating"] + df["Rating_Votes"]) / 4
```

```python
df.sort_values(by = "Final_Rating", ascending=False)
```

## Bonus Challenge. Freaky bonus

```python
def sum_numbers(column, revenue, votes):
    nums = []
    for i in column:
        if i.isdigit() == True:
            nums.append(int(i))
    if sum(nums) >= revenue and sum(nums) <= votes:
        return f"It has a hidden pattern"
    else:
        return f"There is no hidden pattern here..."
```

```python
df["Hidden_Patterns"] = df.apply(lambda x: sum_numbers(x["Description"], x["Revenue (Millions)"], x["Votes"]), axis = 1)
```

```python
df
```

```python
movies_with_hidden_pattern = df[df["Hidden_Patterns"] == "It has a hidden pattern"]
```
