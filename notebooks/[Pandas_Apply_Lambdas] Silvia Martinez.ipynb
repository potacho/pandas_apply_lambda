## Challenge 1. Using a single argument

```python
def bins(votes):
    if votes >= 0 and votes <= 999:
        return "cat_1"
    elif votes >= 1000 and votes <= 9999:
        return "cat_2"
    elif votes >= 10000 and votes <= 99999:
        return "cat_3"
    elif votes >= 100000 and votes <= 999999:
        return "cat_4"
    elif votes >= 1000000:
        return "cat_5"
        
bins(1000)
```

```python
df["Bin"] = df.apply(lambda x: bins(x["Votes"]), axis = 1)
```

```python
df
```

## Challenge 2. Using 2 arguments

```python
def revenue_minute(revenue, minute):
    return revenue / minute
```

```python
df["Revenue per Minute"] = df.apply(lambda x: revenue_minute(x["Revenue (Millions)"], x["Runtime (Minutes)"]), axis = 1)
```

```python
df
```

## Challenge 3. A bit more complicated

```python
df
```

```python
def new_rating(genre, rating):
    if "thriller" in genre.lower():
        rating = rating + 1
    elif "comedy" in genre.lower():
        rating = rating -1
    return rating
```

```python
df["New Rating"] = df.apply(lambda x: new_rating(x["Genre"], x["Rating"]), axis = 1)
```
